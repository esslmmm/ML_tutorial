```python
import numpy as np
import matplotlib.pylab as plt
from sklearn.linear_model import LinearRegression
```

```python
rng = np.random
```

```python
#mock data
x = rng.rand(50)*10
y = 2*x+rng.randn(50)
```

```python
#linear regression model
model = LinearRegression()
```

```python
print(x) #Model 1D
x_new = x.reshape(-1,1)
print(x_new) #Model 2D, model need 2D
```

```python
#train
model.fit(x_new,y)
```

```python
#evaluation model
print("R-SQuare : ",model.score(x_new,y))
print("Intercept : ",model.intercept_)
print("Coefficient : ", model.coef_) #slope
```

```python
#test
x_test = np.linspace(-1,11)
print(x_test.shape)
print(x_test)
x_test_new = x_test.reshape(-1,1) # -1 :automatically rows , 1 : 1 column 
print(x_test_new.shape)
print(x_test_new)

y_test = model.predict(x_test_new)
```

```python
#analysis model & result
plt.scatter(x,y)
plt.plot(x_test,y_test,"-r")
plt.show()
```
