```python
import kagglehub
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
```

```python
# Download latest version
path = kagglehub.dataset_download("uciml/adult-census-income")

df = pd.read_csv(f"{path}/adult.csv")
```

```python
df.head()
```

```python
df.dtypes
```

```python
def cleandata(dataset):
    for column in dataset.columns:
        if dataset[column].dtype == type(object):
            le = LabelEncoder()
            dataset[column]=le.fit_transform(dataset[column])
    return dataset
```

```python
def split_feature_class(dataset, feature):
    features = dataset.drop(feature,axis=1)
    label = dataset[feature].copy()
    return features, label
```

```python
df = cleandata(df)
df
```

```python
#split train, test
train_set, test_set = train_test_split(df,test_size=0.2)
```

```python
#train set
train_features, train_label = split_feature_class(train_set,"income")
```

```python
#test set
test_features, test_label = split_feature_class(test_set,"income")
```

```python
model = GaussianNB()
```

```python
model.fit(train_features,train_label)
```

```python
pred_label = model.predict(test_features)
```

```python
#Accuracy Score
print("Accuracy = ", accuracy_score(test_label,pred_label))
```
