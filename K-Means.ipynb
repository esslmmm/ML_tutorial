```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
x,y = make_blobs(n_samples=300,centers=4,cluster_std=0.5,random_state=0)
```

```python
x
```

```python
print("First Column (X axis)",x[:,0])
# print("Second Column (Y axis)",x[:,1])
```

```python
plt.scatter(x[:,0],x[:,1])#x,y
plt.show()
```

```python
model = KMeans(n_clusters=4)
```

```python
model.fit(x)
```

```python
centers = model.cluster_centers_
print(centers)
```

```python
y_pred = model.predict(x)
print(y_pred)
```

```python
# plt.scatter(x[:,0],x[:,1])#x,y
plt.scatter(x[:,0],x[:,1], c=y_pred)
plt.scatter(centers[0,0],centers[0,1], c='green',label="Centroid 1")
plt.scatter(centers[1,0],centers[1,1], c='blue',label="Centroid 1")
plt.scatter(centers[2,0],centers[2,1], c='red',label="Centroid 1")
plt.scatter(centers[3,0],centers[3,1], c='orange',label="Centroid 1")
plt.legend(frameon=True)
plt.show()
```

```python
x_test,y_test = make_blobs(n_samples=10,centers=4,cluster_std=0.5,random_state=0)

y_pred_test = model.predict(x_test)
```

```python
# plt.scatter(x[:,0],x[:,1])#x,y
plt.scatter(x[:,0],x[:,1], c=y_pred)
plt.scatter(x_test[:,0],x_test[:,1], c=y_pred_test, s=120)
plt.scatter(centers[0,0],centers[0,1], c='green',label="Centroid 1")
plt.scatter(centers[1,0],centers[1,1], c='blue',label="Centroid 1")
plt.scatter(centers[2,0],centers[2,1], c='red',label="Centroid 1")
plt.scatter(centers[3,0],centers[3,1], c='orange',label="Centroid 1")
plt.legend(frameon=True)
plt.show()
```
